```python
#Price distribution

plt.hist(data['Price'],bins=30,color='skyblue')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Average rating distribution

plt.hist(data['Avg ratings'],bins=20,color='orange')
plt.title('Average Ratings Distribution')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.show()
```

```python
#Top 10 Restaurants

top_restaurants=data['Restaurant'].value_counts().head(10)
plt.barh(top_restaurants.index, top_restaurants.values, color='green')
plt.xlabel('Number of occurances')
plt.ylabel('Restaurant')
plt.gca().invert_yaxis()  #highest at top
plt.show()
```

```python
#City wise average price

city_avg_price=data.groupby('City')['Price'].mean().sort_values()
plt.figure(figsize=(10,5))
plt.bar(city_avg_price.index,city_avg_price.values,color='teal')
plt.title('Average Price by City')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
